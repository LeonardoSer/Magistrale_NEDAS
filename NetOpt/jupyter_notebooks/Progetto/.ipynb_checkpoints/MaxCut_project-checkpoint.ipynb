We consider the **Max-Cut** problem on an undirected graph with n nodes and m edges.

$$G = (V , E); \space\space |V| = n; \space\space; |E| = m$$

We investigate a linear size MIP formulation, referred to as **MIP-MaxCut**.

**Branch-and-bound based approaches** for Max-Cut usually use the **semi-metric polytope** which has two formulations:
1. **exponential size formulation** consisting of the **cycle inequalities** 
2. **compact size formulation** consisting of **O(mn) triangle inequalities**. 

However, optimizing over the semi-metric polytope can be computationally demanding due to_
- the slow convergence of **cutting-plane algorithms**;
- the **high degeneracy of formulations based on the triangle inequalities**. 

We will exhibit new structural properties of **MIP-MaxCut** that link the binary variables with the cycle inequalities. 
By fixing a binary variable at 0 or 1 in **MIP-MaxCut** imposes the **integrity of several original variables** and the **satisfaction of a possibly exponential number of cycle inequalities in the semimetric formulation**.

Numerical results show that for sparse instances of Max-Cut, **this approach outperforms the branch-and-cut algorithms based on semi-metric polytope** 

# 1.1 The Max-Cut problem

Let **G = (V,E)** be an undirected graph with **n = |V|** and **m = |E|**. 
Suppose that edges of **G** are weighted by a vector $c \in R^E$.
Denote by $ij$ the edge between the two nodes $i$ and $j$ of $V$ . 

In particular, G could be the complete graph $K_n = (V_n , E_n )$ of n nodes.


A cut in **G**, denoted $\delta(S)$, is the set of the edges that have exactly one end-node in the node subset $S \in V$  . 

For each cut $\delta (S)$, the incidence vector associated with it is a vector  $\mathbb X (\delta (S)) \in \{ 0, 1 \}^E $ where a variable $x_{ij} \in \mathbb X (\delta (S))$ is such that:

$$
x_{ij} = \begin{cases} 1 \text{, if } ij \in \delta (S)\\
0 \text{,    otherwise}
\end{cases}
$$

<center><img src="1.jpg" width="500" /></center>




The **Max-Cut** problem is to find a cut of maximum total cost is equivalently to find a node subset **S** that maximize the following

$$
{max}_{_{ \mathbb X (\delta (S)) \space \forall S \in V } } \sum_{ {ij} \in E } c_{ij}x_{ij}
$$

Hence, **finding a maximum cost** cut is **equivalent to optimizing over the cut polytope CUTP(G)** which is the **convex hull of the incidence vectors associated with the cuts in G**.

For example given the graph **G** in the next figure, it's **CUTP(G)** can be calculated starting from the **points in 3-Dimensional space** given by the vectors $\mathbb X (\delta (S))$, as it's showed in the following code cells.

<center><img src="2.jpg" width="500" /></center>


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
from mpl_toolkits.mplot3d import Axes3D
```

```python
points = np.array([[1,0,1],
                   [1,1,0],
                   [0,1,1],
                   [0,0,0]])
hull = ConvexHull(points)
```

```python
fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(111, projection="3d")

# Plot defining corner points
ax.plot(points.T[0], points.T[1], points.T[2], "ko")

# 12 = 2 * 6 faces are the simplices (2 simplices per square face)
for s in hull.simplices:
    s = np.append(s, s[0])  # Here we cycle back to the first coordinate
    ax.plot(points[s, 0], points[s, 1], points[s, 2], "r-")

# Make axis label
for i in ["x", "y", "z"]:
    eval("ax.set_{:s}label('{:s}')".format(i, i))

plt.show()

# A facet of a polytope of dimension n is a face that has dimension n − 1 (in our case will be triangles)
facets = hull.simplices # facets delimitig the polytope CUTP(G)

for i in range(len(facets)):
    facet = facets[i]
    print("- Vertexes of face", i, "are:", points[facet[0]], points[facet[1]], points[facet[2]], "")
    i += 1
```

## 1.2 Cycle inequalities and the semi-metric polytope


Let $\mathbb C$ be the set of chordless cycles in **G**, where a **chordless cycle C** in G s a cycle whose induced subgraph is the cycle itself. 

<center><img src="3.jpg" width="500" /></center>

For a vector $x \in R^E$ and for any subset $F \subseteq E$, let $y(F)=\sum_{{ij}\in F}{x_{ij}}$.

**FACT:** A cut always intersects a cycle **C** in an even number of edges.

<center><img src="4.jpg" width="800" /></center>

Given the above fact, the **incidence vectors** associated with the cuts in **G** satisfy the following **cycle inequalities** introduced by Barahona and Mahjoub:

$$
y(F) - y({C} \setminus {F}) \leq |F|-1 , \space\space\space\forall C\in \mathbb C  \space\text{and} \space F \subseteq C \space \text{with} \space |F| \space \text{odd} \\
0 \leq x_e \leq 1, \space\space \forall e \in E \space \text{s.t.} \space e \space \text{doesn't belong to any triangle} 
$$

The cycle inequalities together with the illustrated trivial lower bound and upper bound inequalities define the so called **semi-metric polytope** **METP(G)** associated with **G** in $R^E$.

By the trivial inequalities we have realized that the **semi-metric polytope METP(G) is a relaxation of CUTP(G)**, if we then replace the above trivial inequalities by $x\in \{0,1\}^E$ (removing in this way the linear relaxation), we obtain the following **integer formulation for CUTP(G)**

$$
y(F) - y({C} \setminus {F}) \leq |F|-1 , \space\space\space\forall C\in \mathbb C  \space\text{and} \space F \subseteq C \space \text{with} \space |F| \space \text{odd}\\ 
x\in \{0,1\}^E
$$. 

*An illustration of the cycle inequalities with 0/1 constraints is represented in the following image*

<center><img src="5.jpg" width="800" /></center>

Thus, having reduced the **METP(G)** problem to the **CUTP(G)** , we can conclude that is also **linear programming relaxation** for **the Max-Cut problem**.

Note that since there is a priori no known polynomial upper bound (in terms of n and m) on the number of chordless cycles and there may be also an exponential number of possible choices for the set **F** given a chordless cycle C, the above formulation of **METP(G)** has a priori an exponential number of inequalities. 

However, **METP(G)** has polynomial size extended formulations, called **METP(Kn)**, which consist of $O(n^2)$ variables where additional variables correspond to the additional edges which complete **G** to $K_n$.

These extended formulations involve the following so-called **triangle inequalities**:

$$
x_{ij}+x_{ik}+x_{jk} \leq 2, \space\space\space \forall i,j,k\in T \\
x_{ij}-x_{ik}-x_{jk} \leq 0, \\
x_{ik}-x_{ij}-x_{jk} \leq 0, \\
x_{jk}-x_{ij}-x_{ik} \leq 0, \space\space\space \forall i,j,k\in T
$$

where **T** is the set of all (unordered) triples of distinct nodes $i,j,k \in V$ such that at least $ij,ik \space \text{or} \space jk$ is an edge in **E**.

```python
import networkx as nx
import gurobipy as gb
from gurobipy import GRB
import os
import random
import matplotlib.pyplot as plt
import pygraphviz as pygv
import itertools
import numpy as np
from IPython.display import SVG, display
from itertools import combinations, chain, groupby
```

```python
# Drawing Function
def draw(graph, node_color, edge_color):
    G = graph.copy()
    pos = nx.shell_layout(G)
    edge_lab = {(u,v):G[u][v]['cost'] for u,v in G.edges()}
    node_lab = {i for i in G.nodes()}
    plt.figure(figsize=(10,10))
    nx.draw_networkx_edge_labels (G, pos, edge_labels=edge_lab)
    nx.draw(G, pos, with_labels=True, node_color=node_color, edge_color=edge_color, node_size=500)

# Another Drawing Function
def drawSol(graph, node_color, edge_color, x, cut):
    G = graph.copy()
    for i,j in G.edges():
        if(cut):
            if(x[i,j].x != 1):
                G.remove_edge(i,j)
        else:
            if(x[i,j].x == 1):
                G.remove_edge(i,j)
    pos = nx.shell_layout(G)
    edge_lab = {(u,v):G[u][v]['cost'] for u,v in G.edges()}
    node_lab = {i for i in G.nodes()}
    plt.figure(figsize=(10,10))
    nx.draw_networkx_edge_labels (G, pos, edge_labels=edge_lab)
    nx.draw(G, pos, with_labels=True, node_color=node_color, edge_color=edge_color, node_size=500)
```

```python
# returns a random complete graph G which edges are generated with probability p
def gnp_random_connected_graph(n, p):
    edges = combinations(range(n), 2)
    G = nx.Graph()
    G.add_nodes_from(range(n))
    if p <= 0:
        return G
    if p >= 1:
        return nx.complete_graph(n, create_using=G)
    for _, node_edges in groupby(edges, key=lambda x: x[0]):
        node_edges = list(node_edges)
        random_edge = random.choice(node_edges)
        G.add_edge(*random_edge)
        for e in node_edges:
            if random.random() < p:
                G.add_edge(*e)
    return G
```

```python
# return a complete version of graph G with 0-cost edges where missing
def complete_graph(G):
    Kn = G
    
    for i in Kn.nodes():
        for j in Kn.nodes():
            if(i!=j):
                if not ((i,j) in Kn.edges() or (j,i) in Kn.edges()):
                    Kn.add_edge(i, j)
                    Kn[i][j]["cost"] = 0
    return Kn;
```

```python
# A method that given a graph return all chordless cycles of it
def get_chordless_cycles(G):
    cycles = find_all_cycles(G)
    chordless_cycles = []
    
    for cycle in cycles:
        
        cycle_edges = []
        for i in range(len(cycle)-1):
                cycle_edges.append([cycle[i], cycle[i+1]])
        cycle_edges.append([cycle[0], cycle[len(cycle)-1]])
    
        possible_chords = []
        is_chordless = True

        for i in cycle:
            for j in cycle:
                if(i!=j):
                    if(([i,j] not in cycle_edges) and([j,i] not in cycle_edges)):
                        if([i,j] in Kn.edges()):
                            is_chordless = False
         
        if(is_chordless):
            chordless_cycles.append(cycle_edges)
                    
    return chordless_cycles

# A method that given a graph return all cycles of it
def find_all_cycles(G, source=None, cycle_length_limit=None):
    """forked from networkx dfs_edges function. Assumes nodes are integers, or at least
    types which work with min() and > ."""
    if source is None:
        # produce edges for all components
        nodes=[list(i)[0] for i in nx.connected_components(G)]
    else:
        # produce edges for components with source
        nodes=[source]
    # extra variables for cycle detection:
    cycle_stack = []
    output_cycles = set()
    
    def get_hashable_cycle(cycle):
        """cycle as a tuple in a deterministic order."""
        m = min(cycle)
        mi = cycle.index(m)
        mi_plus_1 = mi + 1 if mi < len(cycle) - 1 else 0
        if cycle[mi-1] > cycle[mi_plus_1]:
            result = cycle[mi:] + cycle[:mi]
        else:
            result = list(reversed(cycle[:mi_plus_1])) + list(reversed(cycle[mi_plus_1:]))
        return tuple(result)
    
    for start in nodes:
        if start in cycle_stack:
            continue
        cycle_stack.append(start)
        
        stack = [(start,iter(G[start]))]
        while stack:
            parent,children = stack[-1]
            try:
                child = next(children)
                
                if child not in cycle_stack:
                    cycle_stack.append(child)
                    stack.append((child,iter(G[child])))
                else:
                    i = cycle_stack.index(child)
                    if i < len(cycle_stack) - 2: 
                      output_cycles.add(get_hashable_cycle(cycle_stack[i:]))
                
            except StopIteration:
                stack.pop()
                cycle_stack.pop()
    
    cycles = [list(i) for i in output_cycles]
    return cycles
```

```python
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

# Max-Cut METP(Kn)
Given $G = (V, E)$ of $|V|=n$ nodes and $|E|=m$ arcs, where $c$ is the cost vector associated with |E|:

- Let $K_n = (V, E_{K_n} = E+E^0)$  be the **complete graph** of n nodes build from $G$ by generating **0 cost arcs** $ij \in E^0$ where needed to complete it. </br>
- $c^{k_n}$ is the extended cost vector such that ${c_{ij}}^{k_n} = 0 \space \forall \space ij \in E^0 $</br>
- $\mathbb C$ is the set of **chordless cycles** in $K_n$, where an element $C \in \mathbb C$ is the set of arcs of which the chordless cycle is made.
- $F$ is a subset of arcs $F \in E_{K_n}$ 
- $T$ is is the set of all triples of distinct nodes $i,j,k \in V$ such that at least $ij,ik \space \text{or} \space jk$ is an edge **in the original set of arcs** **E**.

From now on we'll work on $K_n$ so, in order to simplify the notation, we will call $E_{K_n}$, $c^{k_n}$, ${c_{ij}}^{k_n}$ respectively $E$, $c$, $c_{ij}$.

#### Variables
$$
x_{ij} = \begin{cases}1 \text{ if x_ij is in the cut} \\
0 \text{ otherwise}
\end{cases}
$$

$$
y(F)=\sum_{{ij}\in F}{x_{ij}}
$$

#### Formulation
$$
{max} \sum_{ {ij} \in E } c_{ij} x_{ij}\\
y(F) - y({C} \setminus {F}) \leq |F|-1 , \space\space\space\forall C\in \mathbb C  \space\text{and} \space F \subseteq C \space \text{with} \space
|F| \space \text{odd}\\
x_{ij}+x_{ik}+x_{jk} \leq 2, \space\space\space \forall i,j,k\in T \\
x_{ij}-x_{ik}-x_{jk} \leq 0, \\
x_{ik}-x_{ij}-x_{jk} \leq 0, \\
x_{jk}-x_{ij}-x_{ik} \leq 0, \space\space\space \forall i,j,k\in T \\
x\in \{0,1\}^E \\
$$

```python
# we create a random graph with costs on edges
nodes = 8
probability = 0.3

G = gnp_random_connected_graph(nodes,probability)

for u,v in G.edges():
    G[u][v]["cost"] = random.randint(1,100)

draw(G, "lightblue", "black")
```

```python
# we build Kn from G
Kn = complete_graph(G.copy())            
draw(Kn, "lightblue", "black")
```

```python
# we retrieve all chordless cycles of Kn
chordless_cycles = get_chordless_cycles(Kn)
```

```python
# we retrieve all subset of edges we'll need for cycles inequalities
F = []
c_min_f = []
for c in chordless_cycles:
    for subs in powerset(c):
        tmp_c = c.copy()
        subs = np.asarray(subs).tolist()
        if(len(subs)%2==1 and len(subs)>0):
            F.append(subs)
            for i,j in subs:
                tmp_c.remove([i,j])
            c_min_f.append(tmp_c)
```

```python
# we retrieve all subset of 3 nodes [i, j, k] such that at least ij, ik or jk is an arc in the original graph G 
T = []
for subset in powerset(Kn.nodes()):
    if(len(subset) == 3):
        if((subset[0], subset[1]) in G.edges() or (subset[1], subset[0]) in G.edges() \
           (subset[0], subset[2]) in G.edges() or (subset[2], subset[0]) in G.edges() \
           (subset[1], subset[2]) in G.edges() or (subset[2], subset[1]) in G.edges()):
            T.append(np.asarray(subset))
```

```python
maxCut_model = gb.Model()

## Variables
x = maxCut_model.addVars(Kn.edges() ,ub = 1.0, vtype=gb.GRB.BINARY, name='x') # Decision Variables

## Objective function
obj = gb.quicksum([x[u,v] * Kn[u][v]['cost'] for u,v in Kn.edges()])
maxCut_model.setObjective(obj, GRB.MAXIMIZE)

## Cycles inequalities
for i in range(len(F)):
    y_F = gb.quicksum([x[u,v] for u,v in F[i]])
    y_FminC = gb.quicksum([x[u,v] for u,v in c_min_f[i]])
    
    maxCut_model.addConstr(y_F - y_FminC <= abs(len(F[i]) - 1), "cycleIneq" + str(i))

## Triangle Inequalities
count = 0
for i,j,k in T:
    maxCut_model.addConstr(x[i,j] + x[i,k] + x[j,k] <= 2, "triangleIneq0_" + str(count))
    maxCut_model.addConstr(x[i,j] - x[i,k] - x[j,k] <= 0, "triangleIneq1_" + str(count))
    maxCut_model.addConstr(x[i,k] - x[i,j] - x[j,k] <= 0, "triangleIneq2_" + str(count))
    maxCut_model.addConstr(x[j,k] - x[i,j] - x[i,k] <= 2, "triangleIneq3_" + str(count))
    count+=1

maxCut_model.update()    
maxCut_model.write("maxCut.lp")
```

```python
maxCut_model.optimize()
print("\n" + '-'*100 + "\n")
print("value of the solution: ", maxCut_model.objVal)
print("\n" + '-'*100 + "\n")
```

The **semi-metric polytope** forms the core of the linear programming relaxations in many practical solution procedures for the Max-Cut problem. 

Moreover, when **G** is sparse, it is well known that the corresponding relaxation given by the **semi-metric polytope** is quite good.

In this case, for the Max-Cut problem on large sparse graphs, **branch-and-cut** algorithms based on the integer formulation obtained from the **cycle inequalities** and the **0/1 constraints**, remain the best approaches so far (SDP approaches do not take advantage of sparsity and are currently limited to medium-size instances,typically less than 300 nodes).

Hence, optimizing over the **semi-metric polytope** appears to be of key importance.

This is achieved either via a **cutting-plane approach** (using the polynomial time separation algorithm for the cycle inequalities given before) or by solving the compact extended formulation involving $O(n^2)$ variables and $O(nm)$ constraints.
