## mTSP

>**Given**
>A directed graph $G=(N,A)\;$ and a cost $c_{ij} > 0$ for each arc in $A$
>
>**Find**
> A set of routes for $m$ salesmen who all start from and turn back to a home city (depot, node 1) 
 


### Formulation

Decision variables:
$$
x_{ij} = \begin{cases}1 \text{ if arc $(i, j)$ is in the tour} \\
0 \text{ otherwise}
\end{cases}
$$

Formulation:
$$
\begin{alignat}{3}
& \min \sum_{(i,j) \in A} c_{ij} x_{ij} &\\
\text{s.t.} \;\;\;\;\;&\\
\sum_{j = 2}^{j=n} x_{1j} &= m  \;\; \text{(OUT-R)}\\
\sum_{i = 2}^{i=n} x_{i1} &= m \;\; \text{(IN-R)}\\
\sum_{j \in \delta^+(i)} x_{ij} &= 1 \;\; \forall i \in N, i\not=1 \;\; \text{(FS)}\\
\sum_{j \in \delta^-(i)} x_{ji} &= 1 \;\; \forall i \in N, i\not=1 \;\; \text{(RS)}\\
\sum_{(i,j) \in A(S)} x_{ij} &\le |S| - 1 \;\; \forall S \subset N \setminus \{1\}, |S| \ge 2 \;\; \text{(SEC)}\\
\text {or}\\
\sum_{(i,j) \in \delta^+(S)} x_{ij} &\ge 1 \;\; \forall S \subset N, \{1\} \in S\\
x & \in \{0,1\}^{|A|}
\end{alignat}
$$

## mTSP MTZ constraints

### Formulation

Decision variables:
$$
x_{ij} = \begin{cases}1 \text{ if arc $(i, j)$ is in the tour} \\
0 \text{ otherwise}
\end{cases}
$$

Formulation:
$$
\begin{alignat}{3}
& \min \sum_{(i,j) \in A} c_{ij} x_{ij} &\\
\text{s.t.} \;\;\;\;\;&\\
\sum_{j = 2}^{j=n} x_{1j} &= m  \;\; \text{(OUT-R)}\\
\sum_{i = 2}^{i=n} x_{i1} &= m \;\; \text{(IN-R)}\\
\sum_{j \in \delta^+(i)} x_{ij} &= 1 \;\; \forall i \in N, i\not=1 \;\; \text{(FS)}\\
\sum_{j \in \delta^-(i)} x_{ji} &= 1 \;\; \forall i \in N, i\not=1 \;\; \text{(RS)}\\
u_i - u_j + p x_{ij} &\le p-1 \;\; \forall (i,j) \in A, i,j \not = 1\\
x & \in \{0,1\}^{|A|}
\end{alignat}
$$

## **Ex.1_Q1 ->** 1.	Goods are stored only in node 1 (cost: 15 Euro per unit) and delivers are split in two different tours. 

```python
m = 2
p = 14
```

```python
mtsp_mtz = gb.Model()

## Variables
x = mtsp_mtz.addVars(G.edges(),\
                 obj=[G[i][j]['dist']\
                      for i,j in G.edges()],\
             vtype=gb.GRB.BINARY, name='x')

u = mtsp_mtz.addVars(G.nodes(), obj=0.0, vtype=gb.GRB.CONTINUOUS,\
                     lb=1.0, ub=p, name='u')

## Constraints
mtsp_mtz.addConstr(x.sum(root,'*') == m, name = 'OUT-R')
mtsp_mtz.addConstr(x.sum('*',root) == m, name = 'IN-R')

mtsp_mtz.addConstrs((x.sum(i,'*') == 1 \
                 for i in G.nodes() if i != root), name='FS')

mtsp_mtz.addConstrs((x.sum('*',i) == 1 \
                 for i in G.nodes() if i != root), name='RS')

mtsp_mtz.addConstrs((x[i,j] + x[j,i] <= 1 \
                 for i,j in G.edges() if j > i \
                     and i != root and j != root ), name='SUB2')

u[root].lb = 0
u[root].ub = 0

mtsp_mtz.addConstrs((u[i] - u[j] + p * x[i,j] <= p - 1 \
                     for i in G.nodes() for j in G.nodes()\
                    if i != j and i != root and j != root and G.has_edge(i,j)), name='MTZ')

mtsp_mtz._graph = G
mtsp_mtz._vars = x

mtsp_mtz.update()
mtsp_mtz.write('mtsp_mtz.lp')
```

```python
mtsp_mtz.optimize()

print("\n------------------------------------------------------------------------------\n")
print ("-> Solution", mtsp_mtz.objVal)
print("\n------------------------------------------------------------------------------\n")

DrawSol(G, x)
```

## **Ex1_Q2 ->** Goods are stored in nodes 1 (cost: 15 Euro per unit) and node 15 (cost: 200 Euro per unit).

```python
m = 1
p = 27
wharehouse = 15
```

```python
mtsp_mtz = gb.Model()

## Variables
x1 = mtsp_mtz.addVars(G.edges(),\
                 obj=[G[i][j]['dist']\
                      for i,j in G.edges()],\
             vtype=gb.GRB.BINARY, name='x1')

u1 = mtsp_mtz.addVars(G.nodes(), obj=0.0, vtype=gb.GRB.CONTINUOUS,\
                     lb=1.0, ub=p, name='u')

## Constraints
mtsp_mtz.addConstr(x1.sum(root,'*') == m, name = 'OUT-R')
mtsp_mtz.addConstr(x1.sum('*',root) == m, name = 'IN-R')

mtsp_mtz.addConstrs((x1.sum(i,'*') == 1 \
                 for i in G.nodes() if i != root), name='FS')

mtsp_mtz.addConstrs((x1.sum('*',i) == 1 \
                 for i in G.nodes() if i != root), name='RS')

mtsp_mtz.addConstrs((x1[i,j] + x1[j,i] <= 1 \
                 for i,j in G.edges() if j > i \
                     and i != root and j != root ), name='SUB2')

u1[root].lb = 0
u1[root].ub = 0

# the second wharehouse must be in the middle of the tour
u1[wharehouse].ub = 14
u1[wharehouse].lb = 13

mtsp_mtz.addConstrs((u1[i] - u1[j] + p * x1[i,j] <= p - 1 \
                     for i in G.nodes() for j in G.nodes()\
                    if i != j and i != root and j != root and G.has_edge(i,j)), name='MTZ')

mtsp_mtz._graph = G
mtsp_mtz._vars = x1

mtsp_mtz.update()
mtsp_mtz.write('mtsp_mtz.lp')
```

```python
mtsp_mtz.optimize()

print("\n------------------------------------------------------------------------------\n")
print ("-> Solution", mtsp_mtz.objVal)
print("\n------------------------------------------------------------------------------\n")

DrawSol(G, x1)
```

```python
totalcost = 15*u1[15].x + 100*(len(G.nodes()) - u1[15].x)
print("the storage cost for the solution found is ", totalcost)
```

## **Ex1_Q3_A ->** If arc (12,19) is used then arc (19,21) must be used

We can simply model it into the constraint:  $x_{12,19} - x_{19,21} <= 0$

```python
mtsp_mtz.addConstr(x1[12,19] - x1[19,21] <= 0, name="constrA")

mtsp_mtz.update()
mtsp_mtz.write('mtsp_mtz.lp')
```

```python
mtsp_mtz.optimize()

print("\n------------------------------------------------------------------------------\n")
print ("-> Solution", mtsp_mtz.objVal)
print("\n------------------------------------------------------------------------------\n")

DrawSol(G, x1)
```

```python
totalcost = 15*u1[15].x + 100*(len(G.nodes()) - u1[15].x)
print("the storage cost for the solution A found is ", totalcost)
```

## **Ex1_Q3_B ->** The tour must contain the sequence of nodes: 20 -> 21 -> 5 -> 12 -> 1


```python
# remove the constrain added in last execution
mtsp_mtz.remove(mtsp_mtz.getConstrByName("constrA"))

mtsp_mtz.update()
mtsp_mtz.write('mtsp_mtz.lp')
```

we impose that the path must include the subpath 20->21->5->12->1 by setting the bounds of those arcs

```python
x1[20,21].lb = 1
x1[20,21].ub = 1

x1[21,5].lb = 1
x1[21,5].ub = 1

x1[5,12].lb = 1
x1[5,12].ub = 1

x1[12,1].lb = 1
x1[12,1].ub = 1

mtsp_mtz.update()
mtsp_mtz.write('mtsp_mtz.lp')
```

```python
mtsp_mtz.optimize()

print("\n------------------------------------------------------------------------------\n")
print ("-> Solution", mtsp_mtz.objVal)
print("\n------------------------------------------------------------------------------\n")

DrawSol(G, x1)
```

```python
totalcost = 15*u1[15].x + 100*(len(G.nodes()) - u1[15].x)
print("the storage cost for the solution A found is ", totalcost)
```
