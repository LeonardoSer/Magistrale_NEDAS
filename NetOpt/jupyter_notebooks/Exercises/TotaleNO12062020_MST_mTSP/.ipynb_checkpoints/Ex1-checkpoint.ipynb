#  Minimum Spanning Tree Problem 
## Single commodity flow formulation

$$
x_e = \begin{cases}
1 \text{ if edge $e$ is in the tree} \\
0 \text { otherwise}
\end{cases}
$$

Associate to each edge $e$ two directed arcs $(i,j)$ and $(j,i)$ and let 

$$
f_{ij} = \{\text{flow carried by the arc} (i,j)\}
$$

\begin{alignat}{3}
\min cx &&\\
\sum_{e \in E} x_e = n - 1 && \\
\sum_{j \in \delta^+(0)} f_{0j} - \sum_{j \in \delta^-(0)} f_{j0} = n - 1 &&\\
\sum_{j \in \delta^-(v)} f_{jv} - \sum_{j \in \delta^+(v)} f_{vj} = 1 &&\;\; \forall v \in V, v \not = \{0\} \\
f_{ij} \le (n-1)x_e &&\;\; \forall e \in E, e=\{i,j\}\\
f_{ji} \le (n-1)x_e &&\;\; \forall e \in E, e=\{i,j\}\\
f \ge 0, \; x_e \in \{0,1\}
\end{alignat}

```python
args.filename = "net12062020.gml"
args.scale = 20
args.figsize = (15,15)

basename = os.path.splitext(args.filename)[0]
G = nx.read_graphml (args.filename, node_type=int)

print ("G has", G.number_of_nodes(), "nodes and", G.number_of_edges(), "edges")
```

```python
root = 1
G = nx.convert_node_labels_to_integers(G, first_label=root,label_attribute='old' )
D = G.to_directed() # make a directed copy of G
```

```python
mstflow = gb.Model()

## Variables

x = mstflow.addVars(G.edges(), obj=[G[u][v]['cost'] for u,v in G.edges()],
                    ub = 1.0,
                    vtype=gb.GRB.BINARY,
                    name='x')

f = mstflow.addVars(D.edges(), ub=G.number_of_nodes(), vtype=gb.GRB.CONTINUOUS, name='f')

## constraints

mstflow.addConstr(x.sum() == G.number_of_nodes() - 1, 'Cardinality')

mstflow.addConstr(f.sum(root, '*') - f.sum('*', root) == G.number_of_nodes() - 1, 'Flow_at_root')

# Root is the first nodes the iterator moves one step ahead
for v in G.nodes():
    if v != root:
        name = 'Balance_' + str(v)
        mstflow.addConstr(f.sum('*',v) - f.sum(v,'*') == 1, name=name)
    
for e in G.edges():
    
    h = e[0]
    k = e[1]
        
    name = "Force_f(%d,%d)_x(%d,%d)" % (h, k, e[0], e[1])
    mstflow.addConstr(f[h,k] <= (G.number_of_nodes() - 1) * x[e], name=name)

    name = "Force_f(%d,%d)_x(%d,%d)" % (k, h, e[0], e[1])
    mstflow.addConstr(f[k,h] <= (G.number_of_nodes() - 1) * x[e], name=name)

mstflow.update()
mstflow.write('mst.lp')
```

## **Ex1.1 ->**  Node 1 in the final network can accept up to 4 connections

$$ \sum_{e \in \delta(0)} x_e \le \Delta$$ 

```python
rootdeg = 4

mstflow.addConstr(gb.quicksum([x[e] for e in G.edges(root)])  <= rootdeg, name='RootDegree')

mstflow.update()
mstflow.write('mst.lp')

mstflow.optimize()
print("\n------------------------------------------------------------------------------------------------------------------------------------------------\n")
DrawSolFlow(G, x, f)
print("\n------------------------------------------------------------------------------------------------------------------------------------------------\n")
print("object val->", mstflow.objVal)
```

## **Ex1.2 ->** Odd nodes in the final network (but node 1) can accept up to 2 connections while even nodes can accept up to 3 connections

```python
# remove the constraint from Ex1.1
mstflow.remove(mstflow.getConstrByName('RootDegree'))
mstflow.update()
mstflow.write('mst.lp')
```

```python
for i in G.nodes():
    if(i%2 == 0 and i != root):
        mstflow.addConstr(x.sum(i, '*') + x.sum('*', i) <= 3, name="MaxConnections"+str(i))
    elif(i != root):
        mstflow.addConstr(x.sum(i, '*') + x.sum('*', i) <= 2, name="MaxConnections"+str(i))

mstflow.update()
mstflow.write('mst.lp')

mstflow.optimize()

print("\n------------------------------------------------------------------------------------------------------------------------------------------------\n")
print("object val->", mstflow.objVal)
print("\n------------------------------------------------------------------------------------------------------------------------------------------------\n")

DrawSolFlow(G, x, f)
```

## **Ex1.3 ->** The company can install a hub in one of the nodes [12,13,17] with the following features:
 - If the hub is installed in node k, the cost of a link from node k to any other node different from node 1 is zero;
 - The hub can accept a maximum of 6connections.

```python
hub_nodes = [12, 13, 17]
solutions = []

# we find a solution for every possible location of the hub
for curr_hub in hub_nodes:
    
    # read the graph.gml
    args.filename = "net12062020.gml"
    args.scale = 20
    args.figsize = (15,15)

    basename = os.path.splitext(args.filename)[0]
    G = nx.read_graphml (args.filename, node_type=int)

    print ("G has", G.number_of_nodes(), "nodes and", G.number_of_edges(), "edges")
    
    root = 0
    G = nx.convert_node_labels_to_integers(G, first_label=root,label_attribute='old' )
    D = G.to_directed() # make a directed copy of G

    # we set hub's outgoing edges cost to 0
    for i,j in G.edges:
        if(i == curr_hub or j == curr_hub):
            G[i][j]["cost"] = 0
    
    # Instantiate a model
    mstflow = gb.Model()
    
    ## Variables
    x = mstflow.addVars(G.edges(), obj=[G[u][v]['cost'] for u,v in G.edges()],
                        ub = 1.0,
                        vtype=gb.GRB.BINARY,
                        name='x')

    f = mstflow.addVars(D.edges(), ub=G.number_of_nodes(), vtype=gb.GRB.CONTINUOUS, name='f')

    ## constraints

    mstflow.addConstr(x.sum() == G.number_of_nodes() - 1, 'Cardinality')

    mstflow.addConstr(f.sum(root, '*') - f.sum('*', root) == G.number_of_nodes() - 1, 'Flow_at_root')

    # Root is the first nodes the iterator moves one step ahead
    for v in G.nodes():
        if v != root:
            name = 'Balance_' + str(v)
            mstflow.addConstr(f.sum('*',v) - f.sum(v,'*') == 1, name=name)

    for e in G.edges():

        h = e[0]
        k = e[1]

        name = "Force_f(%d,%d)_x(%d,%d)" % (h, k, e[0], e[1])
        mstflow.addConstr(f[h,k] <= (G.number_of_nodes() - 1) * x[e], name=name)

        name = "Force_f(%d,%d)_x(%d,%d)" % (k, h, e[0], e[1])
        mstflow.addConstr(f[k,h] <= (G.number_of_nodes() - 1) * x[e], name=name)
    
    
    ### we add a constraint for hub's connections
    mstflow.addConstr(x.sum(curr_hub, '*') + x.sum('*', curr_hub) <= 6, name="MaxHubConnections")
    
    mstflow.update()
    mstflow.write('mst.lp')
    
    mstflow.optimize()
    solutions.append({"HubNode":curr_hub, "objVal":mstflow.objVal})
    DrawSolFlow(G, x, f)
```

```python
for sol in solutions:    
    print("object val for putting an hub in node", sol["HubNode"], "->", sol["objVal"])
```

### the best location for an hub is **node 13**
